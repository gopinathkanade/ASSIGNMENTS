***MODEL VALIDATION TECHINQUES ***

VIF= Variance Inflamation Factor
for calculation of VIF the R squared of variables should be calculated  

```python
data1.columns
```

```python
rsq_rds=smf.ols('RDS~ADM+MS',data=data1).fit().rsquared
vif_rds=1/(1-rsq_rds)

rsq_adm=smf.ols('ADM~RDS+MS',data=data1).fit().rsquared
vif_adm=1/(1-rsq_adm)

rsq_ms=smf.ols('MS~ADM+RDS',data=data1).fit().rsquared
vif_ms=1/(1-rsq_ms)

# Storing VIF values in dataframe
d1={'VARIABLES':['RDS','ADM','MS'],'VIF':[vif_rds,vif_adm,vif_ms]}
vif_frame=pd.DataFrame(d1)
vif_frame
```

Since No values of VIF is greater than 20 , we can say that there is no collinearity problem. So we shall consider all columns for building the models



***RESIDUAL ANALYSIS ***

**Q-Q PLOT**


```python
model=smf.ols('Profit~ADM+MS+RDS',data=data1).fit()
```

```python
import statsmodels.api as sm
plt.figure(figsize=(90,30))
sns.set_style(style='darkgrid')
qqplot=sm.qqplot(model.resid,line='q')
plt.title('Normal Q-Q plot of residuels')
plt.show()
```

```python
#to print the index of value whose residuels less than -10000
list(np.where(model.resid<-10000))
```

**2.RESIDUAL PLOT FOR HOMOSCEDASTICITY:CONSTANT VARIANCE**

```python
#function defination for standardization i.e.,z score function name:get_standardzed_values
def get_standardized_values (vals):
  return(vals-vals.mean()/vals.std())#this is z-score (Xi-mean/std.dev)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(get_standardized_values(model.fittedvalues),#fitted values are y hat.,model.resid is error
            get_standardized_values(model.resid))

plt.title('Residual plot')
plt.xlabel('Standardized fitted values.i.e.,y hat')
plt.ylabel('Standardizes residual values')
plt.show()
```

**MODEL DELETION DIAGNOSTICS **

**COOK'S DISTANCE **

```python
#load library
from statsmodels.graphics.regressionplots import influence_plot
```

```python
model_influence=model.get_influence() #this method will return 2 values :cooks distance and p values
(c,_)=model_influence.cooks_distance#we dont need p values
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(15,8))
plt.stem(np.arange(len(data1)),np.round(c,3))
plt.xlabel('Row index')
plt.ylabel('Cooks distance')
plt.show()
```

There is no data point whose value is greater than 0.5 Therefore there no outliers/ influencers
