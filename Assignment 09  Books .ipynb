```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder # for transaction
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_csv('/content/book.csv')
df # data preprocessing is not required as data is already transacted
```

▶**Apriori Algorithm**  

**1.Association Rule with 10% support and 70% confidence**

```python
# with 10% support
frequent_items1=apriori(df,min_support=0.1,use_colnames=True)
frequent_items1
```

```python
#with 70% confidence
rules1=association_rules(frequent_items1,metric='lift',min_threshold =0.7)
rules1.sort_values('lift',ascending=False)
```

```python
# visulization of rules1

plt.scatter(rules1['support'],rules1['confidence'],edgecolors='yellow')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```








**2.Association rule with 20% support and 60% confidence**








```python
# 20% support
frequent_items2=apriori(df,min_support=0.2,use_colnames=True)
frequent_items2
```

```python
# 60% confidence
rules2=association_rules(frequent_items2,metric='lift',min_threshold=0.6)
rules2
```

```python
#visulization of rules2

plt.scatter(rules2['support'],rules2['confidence'],edgecolors='magenta',marker='^')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

**3.Association rule with 5% support and 80% confidence**

```python
# with 50 % support
frequent_items3=apriori(df,min_support=0.05,use_colnames=False)
frequent_items3
```

```python
#with 80% confidence
rules3=association_rules(frequent_items3,metric='lift',min_threshold=0.8)
rules3
```

```python
rules3[rules3.lift>1]
```

```python
#visulization of rules3

plt.scatter(rules3['support'],rules3['confidence'],marker='*',edgecolors='maroon')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```
